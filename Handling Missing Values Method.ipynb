# Dropna() Method

```python
df.dropna()                                       # dropna(), Drop/delete All Missing values rows
```

```python
df.dropna(axis=1)       # axis=1, Drop those columns that have missing values, drop age and blood pressure both columns
```

```python
df.dropna(axis=0)                                # Axis=0, Also drop/delete all missing values rows (same as df.dropna())
```

```python
df.dropna(how='all')
```

```python
df.dropna(how='any')
```


# fillna() Method

```python
df.fillna(0)                                               # Fillna(), Fill missing values by 0
```

```python
df.fillna(2)                                               # Fill Missing values by 2
```

```python
df1=pd.read_csv("D:\\Pyhton\\Data\\Data_pre.csv")
df1
```

```python
df1.fillna({'Age':0, 'State':' No State Available', "Diabetic":'None'})   # Fill Missing values by Specific Values/name
```

```python
df1.fillna(method='ffill')                                         # Fill Missing values by previous values/row
```

```python
df1.fillna(method='bfill')                                        # Fill Missing values by next the values/row
```

```python
df1.fillna(method="bfill", axis=0)                            # Fill all missing values by next rows values
```

```python
df1.fillna(method="ffill",axis=0)           # Fill all missing values by previous values              
```

```python
df1.fillna(method="ffill", axis=1)                     # Fill all NaN values by previous columns, bcz of axis 1
```

```python
df1.fillna(method="bfill", axis=1)                          # Fill all NaN/missing values by next columns value
```

```python
df1.fillna(2,limit=4)                      # It will fill 4 values by 2
```

```python
df1.fillna(5000, inplace=True)                 # It will modify the previous data
df1
```
