```python
weather_data={                                        # Second way to Create DaraFrame Using dictionary
    
    'day':['1/1/2017', '1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
    'event':['rain','sunny','snow','snow','rain','sunny'],
    'temperature':[32,35,38,24,32,31],
    'windspeed':[6,7,2,8,7,9]

}
df=pd.DataFrame(weather_data)
df
```

```python
weather=[                                  # Third way to create DataFramde Using Tuple and List
    ('1/1/2017',35,5,'rain'),
    ('1/2/2017',36,4,'sunny'),
    ('1/3/2017',38,8,'snow'),
    ('1/5/2017',25,9,'snow')
]
df1=pd.DataFrame(weather, columns=['day','temp','windspeed','event'])
df1
```
